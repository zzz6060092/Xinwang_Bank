```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_b = pd.read_csv('data_b_train.csv')
train_m = pd.read_csv('data_m_train.csv')
test_b = pd.read_csv('data_b_test.csv')
test_m = pd.read_csv('data_m_test.csv')
y_label = pd.read_csv('y_train.csv')

#显示所有列
pd.set_option('display.max_columns', None)
#显示所有行
pd.set_option('display.max_rows', None)
#设置value的显示长度为100，默认为50
train_m.head(5)
```

```python
ntrain_b = train_b.shape[0]
ntrain_m = train_m.shape[0]
ntest_b = test_b.shape[0]
ntest_m = test_m.shape[0]
all_m = pd.concat((train_m,test_m)).reset_index(drop = True)
```

```python
all_m.drop('x_num_19', axis=1, inplace=True)
all_m.drop('x_num_27', axis=1, inplace=True)
all_m.drop('x_num_38', axis=1, inplace=True)
```

```python
miss_feature = []
for i,x in enumerate(all_m):
    for j in range(all_m.shape[0]):
        if all_m.iat[j,i] == -99:
            if x not in miss_feature:
                miss_feature.append(x)
            break
miss_feature
```

```python
import lightgbm as lgb

from sklearn.model_selection import StratifiedKFold, KFold
def set_missing_feature(train_X,feature):
    ycol = feature
    #将目标属性是否缺失分开
    know_target = pd.DataFrame(train_X[train_X[feature] != -99])
    unknow_target = pd.DataFrame(train_X[train_X[feature] == -99])
    #训练模型
    df_train = know_target
    params = {
        'task': 'train',
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'metric': {'l2', 'auc'},
        'num_leaves': 31,
        'learning_rate': 0.05,
        'feature_fraction': 0.9,
        'bagging_fraction': 0.8,
        'bagging_freq': 5,
        'verbose': 0
        }
    X_train = df_train.drop([ycol],axis = 1)
    Y_train = pd.DataFrame(df_train[ycol])
    lgb_train = lgb.Dataset(X_train,Y_train)

    lgb_eval = lgb.Dataset(X_train,Y_train, reference=lgb_train)
    gbm = lgb.train(params,
                    lgb_train,
                    num_boost_round=20,
                    valid_sets=lgb_eval,
                    early_stopping_rounds=5)

    
    #使用模型进行预测
    predictAge = gbm.predict(unknow_target.drop([ycol],axis = 1), num_iteration=gbm.best_iteration)
    ts = pd.DataFrame(train_X[feature])
    ts[feature].loc[ts[feature] == -99] = predictAge
    return ts
tt = all_m.drop(miss_feature,axis = 1)
for i in range(len(miss_feature)):
    tt = pd.concat([tt,set_missing_feature(all_m.drop(['id', 'timestamp',miss_feature[i^1]],axis = 1),miss_feature[i])],axis = 1)
all_m = tt
all_m.head(5)
```

```python
def cut_bins(raw_data, col_name=None, q=10):
    features, bins = pd.qcut(raw_data[col_name], q=q, retbins=True, duplicates="drop")
    labels = list(range(len(bins) - 1))
    features, bins = pd.qcut(raw_data[col_name], labels=labels, q=q, retbins=True, duplicates="drop")
    return features, bins, labels
features = pd.DataFrame()
for i,x in enumerate(all_m):
    if x == 'id' or x == 'timestamp':
        continue
   # print(x)
    features[x],bins,labels = cut_bins(all_m,x)
```

```python
df_op = pd.concat([all_m['id'],all_m['timestamp'],features],axis = 1)
df_op.head(5)
```

```python
df_op = df_op.astype(str)
```

```python
all_m_to_squence = pd.DataFrame(columns=['id'],data = [_ for _ in range(ntrain_b+ntest_b)])
def sequence(A):
    return list(A)
for i ,x in enumerate(df_op):
    if x=='id' or x=='timestamp':
        continue
    nar =[]
    for j in range(ntrain_b):
        nar.append(sequence(df_op[x][:ntrain_m][all_m['id'] == j]))
    for j in range(ntest_b):
        nar.append(sequence(df_op[x][ntrain_m:][all_m['id'] == j]))
    for i in range(len(nar),ntrain_b+ntest_b):
        nar.append('0')
    all_m_to_squence[x+'_squence'] = nar
all_m_to_squence.to_pickle('all_m_to_squence.pickle')
all_m_to_squence.head(5)
```

```python
from gensim.models import Word2Vec
def emb(df,feature):
    emb_size = 4
    w2v = Word2Vec(df[feature], size=4, window=2, min_count=1, sg=1, iter=5)
    emb_matrix = []
    for seq in df[feature]:
        vec = []
        for w in seq:
            if w in w2v:
                vec.append(w2v[w])
        if len(vec) > 0:
            emb_matrix.append(np.mean(vec, axis=0))
        else:
            emb_matrix.append([0] * emb_size)
    emb_matrix = np.array(emb_matrix)
    global ntrain_b
    global ntest_b
    tmp = pd.DataFrame(columns=['id'],data = [_ for _ in range(ntrain_b+ntest_b)])
    for i in range(emb_size):
        tmp['{}_emb_{}'.format(feature, i)] = emb_matrix[:, i]
    del w2v, emb_matrix
    return tmp.drop(['id'],axis = 1)  
df = pd.DataFrame(columns=['id'],data = [_ for _ in range(ntrain_b+ntest_b)])
for i,x in enumerate(all_m_to_squence):
    if x == 'id':
        continue
    df = pd.concat([df,emb(all_m_to_squence,x)],axis = 1)
```
